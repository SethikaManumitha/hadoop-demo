Columns Standarised

```python
location.columns = location.columns.str.lower().str.replace('[^a-z0-9_]', '', regex=True)
print(location.columns.tolist())
```

```python
weather.columns = weather.columns.str.lower().str.replace(r'\s*\(.*\)', '', regex=True).str.replace('[^a-z0-9_]', '', regex=True)
weather.columns
```

Remove redundant column

```python
weather = weather.drop(columns=['rain_sum'])
print(weather.columns.tolist())
```

Date Standarised

```python
weather['date'] = pd.to_datetime(weather['date']).dt.strftime('%Y-%m-%d')
```

```python
weather.to_csv('processed_weather_data.csv', index=False)
location.to_csv('processed_location_data.csv', index=False)
print("Processed data saved successfully!")
```
