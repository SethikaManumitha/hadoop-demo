```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
weather = pd.read_csv('processed_weather_data.csv')
weather.head()
```

```python
weather.info()
```

```python
weather.describe()
```

```python
features = [
    'location_id',
    'weather_code',
    'temperature_2m_max',
    'temperature_2m_min',
    'temperature_2m_mean',
    'apparent_temperature_max',
    'apparent_temperature_min',
    'apparent_temperature_mean',
    'daylight_duration',
    'sunshine_duration',
    'precipitation_sum',
    'precipitation_hours',
    'wind_speed_10m_max',
    'wind_gusts_10m_max',
    'wind_direction_10m_dominant',
    'shortwave_radiation_sum',
]

target = 'et0_fao_evapotranspiration'

# Create a new list for correlation
corr_columns = features + [target]

corr_matrix = weather[corr_columns].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix of Selected Weather Features')
plt.show()
```

As seen from here features such as shortwave_radiation_sum,sunshine_duration,temperature_2m_max has a strong positive correlation with et0_fao_evapotranspiration while precipitation_hours has a strong negative correlartion

```python
weather['sunshine_duration'].describe()
```

It is also seeing the sunshine_duration seems to have a huge standard variation hinting these data might be skewed with outliers.

```python

# Create a box plot for sunshine_duration
plt.figure(figsize=(8, 6))
plt.boxplot(weather['sunshine_duration'])
plt.ylabel('Sunshine Duration')
plt.title('Box Plot of Sunshine Duration')
plt.grid(True, alpha=0.3)
plt.show()
```

As expected sunshine duration seem to have a huge set of outliers as well as the distribution of the data seem to be skewed

```python
plt.figure(figsize=(10, 6))
plt.hist(weather['sunshine_duration'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Sunshine Duration')
plt.ylabel('Frequency')
plt.title('Distribution of Sunshine Duration')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(weather['et0_fao_evapotranspiration'])
plt.ylabel('ET0 FAO Evapotranspiration')
plt.title('Box Plot of ET0 FAO Evapotranspiration')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
weather['date'] = pd.to_datetime(weather['date'])
weather['month'] = weather['date'].dt.month
may_data = weather[weather['month'] == 5]
```

Decided to only use May data to better fit the task requirement in finding the   expected amount of precipitation_hours, sunshine, and wind_speed that would lead to a lower amount of evapotranspiration for the month of May. Otherwise the predictions might be underfitted

```python
import seaborn as sns
import matplotlib.pyplot as plt

features = ['precipitation_hours', 'sunshine_duration', 'wind_speed_10m_max','shortwave_radiation_sum','temperature_2m_max']
target = 'et0_fao_evapotranspiration'

for f in features:
    plt.figure(figsize=(7,5))
    sns.regplot(data=may_data, x=f, y=target, scatter_kws={'alpha':0.3})
    plt.title(f'{f} vs {target} (Check Linearity)')
    plt.show()
```

What is seeing here is that majority of the features selected show non-linear relationships. Hence a linear regression model might not be suitable
